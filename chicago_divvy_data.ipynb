```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2014 Chicago Divvy Data

```python
# Read in all 2014 rider data
january_2014 = pd.read_csv('Divvy_2014_01.csv')
february_2014 = pd.read_csv('Divvy_2014_02.csv')
march_2014 = pd.read_csv('Divvy_2014_03.csv')
april_2014 = pd.read_csv('Divvy_2014_04.csv')
may_2014 = pd.read_csv('Divvy_2014_05.csv')
june_2014 = pd.read_csv('Divvy_2014_06.csv')
july_2014 = pd.read_csv('Divvy_2014_07.csv')
august_2014 = pd.read_csv('Divvy_2014_08.csv')
september_2014 = pd.read_csv('Divvy_2014_09.csv')
october_2014 = pd.read_csv('Divvy_2014_10.csv')
november_2014 = pd.read_csv('Divvy_2014_11.csv')
december_2014 = pd.read_csv('Divvy_2014_12.csv')

# Merge the 2014 monthly data and sort by trip_id
divvy_2014 = pd.concat([january_2014, february_2014, march_2014, april_2014, may_2014,
                       june_2014, july_2014, august_2014, september_2014,
                       october_2014, november_2014, december_2014])
divvy_2014 = divvy_2014.sort_values(by='starttime').reset_index()
```

## Divvy Ridership by Month and Year
* Ride count by month for each year: 2014, 2015, 2016, 2017
* Comparison between each year

```python
# Monthly and aggregate ride count for 2014
jan_2014_ride_count = january_2014[['trip_id']].count()
feb_2014_ride_count = february_2014[['trip_id']].count()
mar_2014_ride_count = march_2014[['trip_id']].count()
apr_2014_ride_count = april_2014[['trip_id']].count()
may_2014_ride_count = may_2014[['trip_id']].count()
jun_2014_ride_count = june_2014[['trip_id']].count()
jul_2014_ride_count = july_2014[['trip_id']].count()
aug_2014_ride_count = august_2014[['trip_id']].count()
sep_2014_ride_count = september_2014[['trip_id']].count()
oct_2014_ride_count = october_2014[['trip_id']].count()
nov_2014_ride_count = november_2014[['trip_id']].count()
dec_2014_ride_count = december_2014[['trip_id']].count()

# create 2014 monthly ride count dataframe
ride_count_2014 = [jan_2014_ride_count, feb_2014_ride_count, mar_2014_ride_count, apr_2014_ride_count,
                  may_2014_ride_count, jun_2014_ride_count, jul_2014_ride_count, aug_2014_ride_count,
                  sep_2014_ride_count, oct_2014_ride_count, nov_2014_ride_count, dec_2014_ride_count]
monthly_ride_count_2014 = pd.DataFrame(ride_count_2014)
monthly_ride_count_2014['Month']=('January', 'February', 'March', 'April', 'May', 'June',
'July', 'August', 'September', 'October', 'November', 'December')
monthly_ride_count_2014['Year']=('2014')
monthly_ride_count_2014 = monthly_ride_count_2014.rename(columns={'trip_id': 'Ride Count'})
monthly_ride_count_2014 = monthly_ride_count_2014[['Year', 'Month', 'Ride Count']]


# develop a bar chart to visualize the above monthly data
sns.barplot(x=monthly_ride_count_2014['Month'], y=monthly_ride_count_2014["Ride Count"], color='#a0c0c1')
plt.title("2014 Divvy Ride Count by Month")
plt.ylabel("Count of Rides")
plt.xlabel("Months")
plt.xticks(rotation=90)
plt.grid(linestyle='dotted')
plt.tight_layout()
#plt.savefig("2014 Monthly Ride Count.png")

plt.show()
```
